```python
year = terrorism_df['Year'].unique()
years_count = terrorism_df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year, y = years_count, palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=25)
plt.show()
```

```python
attack = terrorism_df.Country.value_counts()[:10]
print(attack)
```

```python
terrorism_df.Group.value_counts()[1:10]
```

```python
plt.subplots(figsize=(15,10))
sns.barplot(terrorism_df['Country'].value_counts()[:10].index,terrorism_df['Country'].value_counts()[:10].values,palette='YlOrBr_r')
plt.title('Top 10 Countries Affected By Terrorism', fontsize=25)
plt.xlabel('Country', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.xticks(rotation = 50)
plt.show()
```

```python
df = terrorism_df[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(15,10))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title("Deaths from Terrorism",fontsize=25)
plt.ylabel("Deaths",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
terrorism_df['Attacktype'].value_counts().plot(kind='bar',figsize=(15,10),color='brown')
plt.xticks(rotation = 30)
plt.xlabel("Type of Attack",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.title("Type of Attacks",fontsize=25)
plt.show()
```

```python
terrorism_df[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkslateblue'])
plt.xticks(rotation=30)
plt.title("Deaths in Various Attacks",fontsize=25)
plt.ylabel('Number of People',fontsize=20)
plt.xlabel('Type of Attack',fontsize=20)
plt.show()
```
